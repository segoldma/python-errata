#### Getting Visitor History

```python
url = "https://app.pendo.io/api/v1/visitor/0004D96F-F3E0-4FC0-B0DA-9336FABB3980/history"

params = {
    "starttime":"1600205888000"
}

headers = {
    'x-pendo-integration-key':os.environ["SETH_PENDO_KEY"],
    'content-type': "application/x-www-form-urlencoded"
}

response = requests.get(url, headers = headers, params = params)
```

#### Format JSON response

```python
visitor_history = response.json()
```

#### Convert to pandas DataFrame

```python
from pandas import json_normalize

df = json_normalize(visitor_history)
```

```python
import pandas as pd

df['timestamp'] = pd.to_datetime(df['ts'], unit='ms')
df['last_timestamp'] = pd.to_datetime(df['lastTs'],unit = 'ms')
```
